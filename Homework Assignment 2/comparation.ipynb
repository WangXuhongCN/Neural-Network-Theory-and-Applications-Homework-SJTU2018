```python
%store -r fpr_m_ran
%store -r tpr_m_ran
%store -r fpr_m_priori
%store -r tpr_m_priori
%store -r fpr_m_svm
%store -r tpr_m_svm
```

```python
roc_auc_ran = auc(fpr_m_ran, tpr_m_ran)
roc_auc_priori = auc(fpr_m_priori, tpr_m_priori)
roc_auc_svm = auc(fpr_m_svm, tpr_m_svm)
```

```python
plt.figure()
lw = 2
plt.plot(fpr_m_svm, tpr_m_svm, color='darkorange',
         lw=lw, label='SVM average ROC curve (area = %0.2f)' % roc_auc_svm)
plt.plot(fpr_m_priori, tpr_m_priori, color='green',
         lw=lw, label='priori M3SVM average ROC curve (area = %0.2f)' % roc_auc_priori)
plt.plot(fpr_m_ran, tpr_m_ran, color='blue',
         lw=lw, label='random M3SVM average ROC curve (area = %0.2f)' % roc_auc_ran)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('average ROC curve of 3 type of SVM')
plt.legend(loc="lower right")
plt.savefig('roc compare svm.png')
sb.plt.show()
```
